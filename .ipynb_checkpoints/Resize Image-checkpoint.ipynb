```python
def convolution_output(n, p, f, s, pad_type="valid"):
    """
    Calculates output height and width of convolution.

    n: height, width size
    p: padding in case of pad_type = 'valid'
    f: filter size
    s: stride
    pad_type: 'valid' or 'same'

    """

    if pad_type == "valid":
        return floor((n + 2 * p - f) / s) + 1
    elif pad_type == "same":
        # p = max(f - s, 0) if n % s == 0 else max(f - (n % s), 0)
        # return floor((n + 2 * p - f) / s) + 1
        return ceil(float(n) / float(s))
```

```python
def convolution_transpose_output(n, p, f, s, pad_type="valid"):
    """
    Calculates output height and width of convolution transpose.

    n: height, width size
    p: padding in case of pad_type = 'valid'
    f: filter size
    s: stride
    pad_type: 'valid' or 'same'

    """

    # These statements are from official tensorflow code...
    if pad_type == "valid":
        return n * s + max(f - s, 0)
    elif pad_type == "same":
        return n * s
```

```python
import tensorflow as tf
from math import floor, ceil
```

```python
n = 28
f = 3
s = 3
p = 0

x = tf.random.normal(shape=(1, n, n, 3))
```

```python
conv2d = tf.keras.layers.Conv2D(
    64, kernel_size=f, strides=s, activation="relu", padding="valid"
)
output = conv2d(x)
print(output.shape)
print(convolution_output(n, p, f, s))
```

```python
conv2d = tf.keras.layers.Conv2D(
    64, kernel_size=f, strides=s, activation="relu", padding="same"
)
output = conv2d(x)
print(output.shape)
print(convolution_output(n, p, f, s, pad_type="same"))
```

```python
n = 28
f = 5
s = 3
p = 0

x = tf.random.normal(shape=(1, n, n, 3))
```

```python
conv2d_transpose = tf.keras.layers.Conv2DTranspose(
    64, kernel_size=f, strides=s, activation="relu", padding="valid"
)
output = conv2d_transpose(x)
print(output.shape)
print(convolution_transpose_output(n, p, f, s))
```

```python
conv2d_transpose = tf.keras.layers.Conv2DTranspose(
    64, kernel_size=f, strides=s, activation="relu", padding="same"
)
output = conv2d_transpose(x)
print(output.shape)
print(convolution_transpose_output(n, p, f, s, pad_type="same"))
```
